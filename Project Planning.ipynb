```r
library(tidyverse)
players <- read_csv("players.csv")
players
```

The above "players" data set has 196 observations and 7 variables, so it contains data on 196 unique players. 

The variables in the dataset are:
- `experience` - a categorical variable that represents the level of experience of each player
- `subscribe` - a categorical variable that shows if the player is subscribed to the newsletter
- `hashedEmail` - a categorical variable that shows the (partially hidden) email of each player
- `played_hours` - a numerical variable that represents the number of playing hours
- `name` - a categorical variable that shows the name of each player
- `gender` - a categorical variable that shows the gender of each player
- `Age` - a numerical variable that shows the age of each player

```r
median_hours <- players |>
                    summarise(median_played_hrs = median(played_hours))
median_hours
median_age <- players |>
                summarise(median_age = median(Age, na.rm = TRUE))
median_age
```

Summary Statistics: 
- `median_played_hrs` = 0.1 hours
- `median_age` = age 19 

Issues found in the data: 
- There are 2 NAs in the age column; these missing entries can be informative but since there are only 2 it is a minor issue
- A lot of players have 0 playing hours, this could pull down the mean of the plaeyd hours, skeweing the data

Potential Issues: 
- Age and Gender are both self-reported so data could be biased
- The "0"s in `played_hours` could represent two different types of players, one could be they were playing offline or  players never intended to play but odnwloaded the game, or there are individuals that could hav eplayed but for such a short amount of time the system rounded it down. But the data doesn't distinguish between the two


```r
sessions <- read_csv("Project Planning/sessions.csv")
sessions
```

The above "sessions" data set has 1535 observations and 5 variables, so there were 1535 individual playing sessions. 

The variables in the dataset are:
- `hashedEmail` - a categorical variable that shows the (partially hidden) email of each player
- `start_time` - a time variable that shows the specific date and time a session began
- `end_time` - a time variable that shows the specific date and time a session ended
- `original_start_time` - a numerical variable that represents the time a session started after converted to machine-readable format
- `original_end_time` - a numerical variable that represents the time a session ended after converted to machine-readable format

```r
mean_session_summary <- sessions |>
            summarise(mean_duration_min = mean(original_end_time - original_start_time, na.rm = TRUE)/60000)
mean_session_summary
```

Summary Statistics: 
- `mean_duration_min` represents the average session duration in minutes

Issues found in the data: 
- the data table contains 4 time-related columns, two of which are redundant (`start_time` and `end_time`), which are less reliable for calculations

Potential Issues: 
- the end time could be inaccurate of the actual playing time because the player could have just left the game on
