(3) Exploratory Data Analysis and Visualization
- 

To answer the broad question above, only the `original_players` data is necessary. 

```r
original_players <- read_csv("players.csv")
head(original_players)
```

```r
players <- original_players |>
            rename(age = Age) |>
            mutate(experience = as.factor(experience), 
                   gender = as.factor(gender), 
                   subscribe = as.factor(subscribe))
players     
```

```r
mean_values <- players |> summarise(
    mean_played_hours = mean(played_hours, na.rm = TRUE), 
    mean_age = mean(age, na.rm = TRUE))
mean_values
```

```r
experience_hours_summary <- players |>
            group_by(experience) |>
            summarise(mean_hours = mean(played_hours, na.rm = TRUE))
experience_hours_plot <- experience_hours_summary |>
                ggplot(aes(x = experience, y = mean_hours)) + 
                geom_bar(stat = "identity") + 
              labs(
                title = "Average Played Hours by Experience Level", 
                x = "Experience Level", 
                y = "Average Hours Played") + 
            theme(text = element_text(size = 18))
experience_hours_plot
```

The `experience_hours_plot` shows how average played time differs between experience level of players. We see that regular leveled players currently have the highest average hours played, suggesting that experience level may be linked to game engagement, which may matter for predicting whether a player will subscribe to a game-related newsletter. However, we must note that our dataset may be unevenly distributed as in we have fewer players in the veteran category, so the plot may not be fully accurate. 

```r
sub_hours_summary <- players |>
            group_by(subscribe) |>
            summarise(mean_hours = mean(played_hours, na.rm = TRUE))
sub_hours_plot <- sub_hours_summary |>
                ggplot(aes(x = subscribe, y = mean_hours)) + 
                geom_bar(stat = "identity") + 
              labs(
                title = "Average Played Hours by Subscription Status", 
                x = "Subscription Status\n (True / False)", 
                y = "Average Hours Played") + 
            theme(text = element_text(size = 18))
sub_hours_plot
```

The `sub_hours_plot` shows how the average hours played differs by subcription status. We can observe that subscribed players, on average, tend to play more hours than non-subscribed players, suggesting a possible positive relationship between game engagement and the likelihood of subscribing to the newsletter. 

(4) Methods and Plan
- 

I will use a KNN classification model. This is a classification question because we are predicting the label of a categorical variable (Subscription status).

This method is appropriate because it is flexible and needs few assumptions about what the data must look like. Moreover, since KNN predicts based on the outcomes of the most similar players (nearest neighbors), this method is appropriate because players with similar characteristics (hours played) are likely to have similar subscription behavior. 

The weaknesses, though, is that KNN is very slow as the training data gets larger, it is also sensitive to scale, as well as struggle with noisy data.
